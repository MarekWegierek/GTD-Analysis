### 0) Loading and first data cleaning

```python
import pandas as pd
from geopy.geocoders import Nominatim
pd.set_option('display.max_columns', None) # required to see all 135 columns in dataset
```

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv', 
                 encoding_errors='ignore',
                low_memory=False)
```

```python
df.head()
```

#### 1.1) According to the Codebook, if it is impossible to assert the correct date to the case of attack, months are declared as '0'. 
#### Let's check these cases

```python
df['imonth'].value_counts().sort_values(ascending=True, 
                                        ignore_index=True)
```

#### 1.2) As we see, there is 20 of such cases. Lets check the information about them

```python
df.loc[df['imonth'] == 0]
```

#### 1.3) All of these cases lack a lot of information, and giving the fact how little of them there is, we can safely discard them.

```python
removed_rows = pd.DataFrame(data=df.loc[df['imonth'] == 0])
removed_rows['Reason_of_rejection'] = 'Inacurate date'
df = df.loc[df['imonth'] != 0]
```

#### Zaczniemy od oczyszczania wydarze, do ktrych nie moemy przypisalokalizacji

```python
lacking_coordinates = pd.DataFrame(data=
                                   df.loc[(
                                        (df['latitude'].isna()) |
                                        (df['longitude'].isna())
                                        )])

lacking_coordinates['Reason_of_rejection'] = 'Lack of coordinates'
removed_rows = pd.merge(removed_rows, lacking_coordinates, how='outer')
df = df.loc[(
    (~df['latitude'].isna()) |
    (~df['longitude'].isna())
)]
```

```python
removed_rows['Reason_of_rejection'].value_counts()
```

#### inicjalizujemy modu geopy

```python
geolocator = Nominatim(user_agent='GTD Analysis')
```

```python
def checklocal(Latitude, Longitude):
    return geolocator.reverse(Latitude+", "+Longitude, language='en')
```

```python
list2 = df.loc[df['country'] == 236, ['eventid','latitude', 'longitude']].values.tolist()
```

```python
for x in list2:
    print(str(
        checklocal(str(x[1]),str(x[2])))
          .split(',')[-1])
```
